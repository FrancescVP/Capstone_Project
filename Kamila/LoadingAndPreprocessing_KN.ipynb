```python
from scipy.stats import shapiro
#creating empty dataframes to store 
Patients_Gaussian = pd.DataFrame(columns=['ROI', 'Shapiro_stat', 'pvalue'])
Patients_NonGaussian = pd.DataFrame(columns=['ROI', 'Shapiro_stat', 'pvalue'])
Controls_Gaussian = pd.DataFrame(columns=['ROI', 'Shapiro_stat', 'pvalue'])
Controls_NonGaussian = pd.DataFrame(columns=['ROI', 'Shapiro_stat', 'pvalue'])     
#Statistical tests to check if the data has a Gaussian distribution.
feats = fa.columns.to_list()
feats
```

```python
for attr in feats:
    stat,p = shapiro(patients_fa[attr])
    print(attr,'stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        Patients_Gaussian = Patients_Gaussian.append({'ROI': attr, 'Shapiro_stat': stat, 'pvalue': p}, ignore_index=True)

        #print(attr, 'Probably Gaussian')
        
    else:
        Patients_NonGaussian = Patients_NonGaussian.append({'ROI': attr, 'Shapiro_stat': stat, 'pvalue': p}, ignore_index=True)

        #print(attr,'Probably not Gaussian')

        
```

```python
for attr in feats:
    stat,p = shapiro(controls_fa[attr])
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        Controls_Gaussian = Controls_Gaussian.append({'ROI': attr, 'Shapiro_stat': stat, 'pvalue': p}, ignore_index=True)
        #print(attr, 'Probably Gaussian')
        
    else:
        Controls_NonGaussian = Controls_NonGaussian.append({'ROI': attr, 'Shapiro_stat': stat, 'pvalue': p}, ignore_index=True)
        #print(attr,'Probably not Gaussian')
```

 #### Performing t-tests          


```python
from scipy.stats import ttest_ind
t_tests = pd.DataFrame(columns=['ROI', 't', 'pvalue'])
controls = controls_fa.iloc[:,:-4]
patients = patients_fa.iloc[:,:-4]
feats_brain = [e for e in feats if e not in (('age',), ('sex',), ('dd',), ('edss',))]
for attr in feats_brain:
    stat,p = ttest_ind(patients[attr], controls[attr])
    #print(attr, ': stat=%.3f, p=%.3f' % (stat, p))
    t_tests = t_tests.append({'ROI': attr, 't': stat, 'pvalue': p}, ignore_index=True)
```

### Correction for multiple comparisons

```python
len(t_tests)
```

```python
from statsmodels.stats.multitest import fdrcorrection
```

```python
 test, p_corr = fdrcorrection(t_tests["pvalue"], alpha = 0.05, method = "indep", is_sorted = False)
```

```python
t_tests ["Rejected"] = test
t_tests ["p_corr"] = p_corr

print("T-tests uncorrected p<.05:", t_tests[t_tests["pvalue"] < 0.05].count()["t"],"\n",
      "T-tests uncorrected p<.01:", t_tests[t_tests["pvalue"] < 0.01].count()["t"],"\n",
      "T-tests corrected p<.05:",t_tests[t_tests["Rejected"]==True].count()["Rejected"],"\n",
      "T-tests corrected p<.01:", t_tests[t_tests["p_corr"] < 0.01].count()["t"])
```

### Spearman correlation

```python
#selecting columns with connections that differentiate between controls and patients
difference_fa = t_tests[t_tests["p_corr"] < 0.01]
different_fa = difference_fa["ROI"].tolist()
len(different_fa)
```

```python
fa_for_corr = fa[different_fa]
```

```python
fa_for_corr.head()
```

```python
from scipy.stats import spearmanr
spearman = spearmanr(fa_for_corr)
rho, spearman_p = spearman.correlation,spearman.pvalue
```

```python
fa_spearman_rho = pd.DataFrame(rho, columns = different_fa, index = different_fa)
fa_spearman_p = pd.DataFrame(spearman_p, columns = different_fa, index = different_fa)
```
